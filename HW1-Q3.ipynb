```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x1 = np.random.normal(1,0.5,1000)
y1 = np.random.normal(1,0.5,1000)
x2 = np.random.normal(-1,0.5,10)
y2 = np.random.normal(-1,0.5,10)
```

```python
plt.scatter(x1, y1, c='g')
```

```python
plt.scatter(x2, y2, c='g')
```

```python
one1 = np.ones(1000)
one2 = np.ones(10)
neg_one2 = -1 * one2

data1 = np.c_[x1,y1,one1]
data2 = np.c_[x2,y2,neg_one2]
data_1 = np.concatenate((data1,data2))
```

```python
plt.grid()
for i in range(len(data_1)) :
    point = data_1[i]

    if point[2] == 1 :
        color = "g"
    if point[2] == -1 :
        color = "r"
        
    plt.scatter(point[0], point[1], c=color)
```

```python
x1 = np.random.normal(1,0.5,100)
y1 = np.random.normal(1,0.5,100)
x2 = np.random.normal(-1,0.5,100)
y2 = np.random.normal(-1,0.5,100)
one1 = np.ones(100)
one2 = np.ones(100)
neg_one2 = -1 * one2

data1 = np.c_[x1,y1,one1]
data2 = np.c_[x2,y2,neg_one2]
data_2 = np.concatenate((data1,data2))
plt.grid()
for i in range(len(data_2)) :
    point = data_2[i]

    if point[2] == 1 :
        color = "g"
    if point[2] == -1 :
        color = "r"
        
    plt.scatter(point[0], point[1], c=color)
```

```python
#for data_1
learning_rate = 0.01

# generating the weights & the bias
np.random.seed(10)
w1 = np.random.randn()
w2 = np.random.randn()
b = np.random.randn()

errors = []
for epoch in range(150) :
    for i in range(130) :
        
        x = data_1[i]
        
        net = x[0] * w1 + x[1] * w2 + b
    
        target = x[2]
    
        ERROR =  target - np.tanh(net)
        
        #error variations
        errors.append((0.5)*np.square(ERROR))
          
        # Updating the weights & the bias
        w1 = w1 + learning_rate * (ERROR) * x[0]
        w2 = w2 + learning_rate * (ERROR) * x[1]
        b = b + learning_rate * (ERROR)
        
#print final weights & bias
print(w1,w2,b)
        
```

```python
plt.grid()
for i in range(len(data_1)) :
    point = data_1[i]

    if point[2] == 1 :
        color = "b"
    if point[2] == -1 :
        color = "r"
        
    plt.scatter(point[0], point[1], c=color)

# plot seperating line
x3 = np.linspace(-3,3,1010)
y3 = ((-w1 * x3) - b) / w2
plt.plot(x3,y3,c='green')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Adaline data classifier')
```
